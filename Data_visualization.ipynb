```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
x = [1,2,3,4,5,6]
y = [1,2,3,4,5,6]
y1 = [5,4,7,8,9,12]
plt.figure(figsize=(7,5))
plt.plot(x,y,color = 'blue',marker='*')
plt.plot(x,y1,color = 'red',marker='o',linestyle='dashed')
plt.grid(True)
plt.xlabel('X axis')
plt.ylabel('y axis')
plt.title('Line plot')
```

```python
plt.scatter(x,y1)
plt.show()
```

```python
plt.bar(x,y1)
plt.show()
```

```python
plt.hist(y1,bins=4)
```

```python
salary = [15000,23000,70000,50000]
label = ['Intern','Employee','senior','manager']

plt.pie(salary,labels = label,colors=['red','yellow','green','blue'],autopct='%2.2f%%')
plt.title('pie chart')
plt.show()
```
